# Transaction Categorization — NLP Classification Model

**Goal:** Classify raw Indian bank transaction text into spending categories.

**Labels:** `Food`, `Shopping`, `Travel`, `EMI`, `Investment`

**Datasets:**
- `data/financial_transaction_train.csv` — 10,000 labeled transactions
- `data/financial_transaction_test.csv` — 1,001 labeled transactions (held-out evaluation)

**Output:** `models/transaction_classifier.pkl`

## 1. Imports & Configuration

````python
import re
import os
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score, cross_validate
from sklearn.metrics import (
    accuracy_score, f1_score, classification_report,
    confusion_matrix, ConfusionMatrixDisplay
)

warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.dpi'] = 120

RANDOM_STATE = 42
DATA_DIR = 'data'
MODEL_DIR = 'models'
os.makedirs(MODEL_DIR, exist_ok=True)

print('All imports loaded successfully.')
````

## 2. Load Data

````python
train_df = pd.read_csv(os.path.join(DATA_DIR, 'financial_transaction_train.csv'))
test_df  = pd.read_csv(os.path.join(DATA_DIR, 'financial_transaction_test.csv'))

print(f'Train shape : {train_df.shape}')
print(f'Test shape  : {test_df.shape}')
print('\nTrain columns:', train_df.columns.tolist())
train_df.head()
````

## 3. Exploratory Data Analysis

````python
# ── Sample texts per category ────────────────────────────────────────────────
for label, group in train_df.groupby('Label'):
    print(f'\n── {label} ──')
    print(group['Transaction_Text'].sample(3, random_state=RANDOM_STATE).to_string(index=False))
````

## 4. Data Cleaning

Each transaction text looks like:
```
Swiggy order payment | Ref:0f8f77bb | Amount: INR 37090.45
```

Cleaning steps:
1. Extract `Amount` as a numeric feature.
2. Strip the `Ref:` code and `Amount:` portion — these are noise for text classification.
3. Lowercase and remove non-alphabetic characters.
4. Drop duplicates and nulls.

> **Note:** There is no `Cash` label in this dataset. The classification labels are purely based on
> merchant/vendor type, which is realistic for a bank statement parser.

````python
AMOUNT_PATTERN = re.compile(r'Amount:\s*INR\s*([\d,]+\.?\d*)', re.IGNORECASE)
REF_PATTERN    = re.compile(r'\|\s*Ref:[a-z0-9]+', re.IGNORECASE)
AMT_TEXT_PAT   = re.compile(r'\|\s*Amount:.*$', re.IGNORECASE)

def extract_amount(text):
    match = AMOUNT_PATTERN.search(text)
    if match:
        return float(match.group(1).replace(',', ''))
    return np.nan

def clean_text(text):
    text = REF_PATTERN.sub('', text)      # remove Ref:xxxx
    text = AMT_TEXT_PAT.sub('', text)     # remove | Amount: INR ...
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)  # keep only letters
    text = re.sub(r'\s+', ' ', text).strip().lower()
    return text

for df in [train_df, test_df]:
    df['amount']       = df['Transaction_Text'].apply(extract_amount)
    df['clean_text']   = df['Transaction_Text'].apply(clean_text)
    df['amount_log']   = np.log1p(df['amount'])  # log-transform for scale

print('Null amounts in train:', train_df['amount'].isna().sum())
print('Null amounts in test :', test_df['amount'].isna().sum())
print('\nCleaned text sample:')
train_df[['Transaction_Text', 'clean_text', 'amount']].head(4)
````

````python
# ── Check class balance after cleaning ──────────────────────────────────────
print('Duplicate rows removed from train:', train_df.duplicated(subset=['Transaction_Text']).sum())
train_df.drop_duplicates(subset=['Transaction_Text'], inplace=True)
train_df.reset_index(drop=True, inplace=True)

print('\nFinal train shape:', train_df.shape)
print('Label distribution:')
print(train_df['Label'].value_counts())
````

## 5. Feature Engineering

We build a **TF-IDF** representation of the cleaned merchant text.
Unigrams + bigrams capture phrases like `"swiggy order"` or `"home loan emi"`.

````python
X_train_text = train_df['clean_text'].values
X_test_text  = test_df['clean_text'].values

y_train = train_df['Label'].values
y_test  = test_df['Label'].values

print('Training samples :', len(X_train_text))
print('Test samples     :', len(X_test_text))
print('Classes          :', sorted(set(y_train)))
````

## 6. Model Training with Cross-Validation

We use `sklearn.Pipeline` so the TF-IDF vectorizer is fitted only on training folds.
This prevents data leakage.

Three models are compared:
| Model | Notes |
|---|---|
| Logistic Regression | Fast baseline, works well with TF-IDF |
| LinearSVC | Strong for high-dim sparse text features |
| Random Forest | Ensemble, captures non-linear patterns |

````python
# ── Define pipelines ─────────────────────────────────────────────────────────
tfidf_params = dict(
    ngram_range=(1, 2),
    max_features=15_000,
    sublinear_tf=True,   # log-scale TF to reduce dominance of frequent terms
    min_df=2
)

pipelines = {
    'Logistic Regression': Pipeline([
        ('tfidf', TfidfVectorizer(**tfidf_params)),
        ('clf',   LogisticRegression(max_iter=1000, C=5.0, random_state=RANDOM_STATE))
    ]),
    'LinearSVC': Pipeline([
        ('tfidf', TfidfVectorizer(**tfidf_params)),
        ('clf',   LinearSVC(max_iter=2000, C=1.0, random_state=RANDOM_STATE))
    ]),
    'Random Forest': Pipeline([
        ('tfidf', TfidfVectorizer(**tfidf_params)),
        ('clf',   RandomForestClassifier(n_estimators=200, n_jobs=-1, random_state=RANDOM_STATE))
    ]),
}

# ── 5-fold Stratified Cross-Validation ──────────────────────────────────────
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

cv_results = {}
print('Running cross-validation (this may take ~1-2 min)...\n')

for name, pipe in pipelines.items():
    scores = cross_validate(
        pipe, X_train_text, y_train, cv=cv,
        scoring=['accuracy', 'f1_macro', 'f1_weighted'],
        n_jobs=-1
    )
    cv_results[name] = {
        'accuracy'    : scores['test_accuracy'].mean(),
        'f1_macro'    : scores['test_f1_macro'].mean(),
        'f1_weighted' : scores['test_f1_weighted'].mean(),
    }
    print(f'{name:25s} | Acc: {cv_results[name]["accuracy"]:.4f} '
          f'| F1-macro: {cv_results[name]["f1_macro"]:.4f} '
          f'| F1-weighted: {cv_results[name]["f1_weighted"]:.4f}')
````

````python
# ── Visualize CV results ─────────────────────────────────────────────────────
cv_df = pd.DataFrame(cv_results).T.reset_index()
cv_df.rename(columns={'index': 'Model'}, inplace=True)

cv_melt = cv_df.melt(id_vars='Model', var_name='Metric', value_name='Score')

fig, ax = plt.subplots(figsize=(11, 4))
sns.barplot(data=cv_melt, x='Model', y='Score', hue='Metric', ax=ax)
ax.set_ylim(0.8, 1.01)
ax.set_title('5-Fold CV Performance Comparison')
ax.legend(loc='lower right')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.3f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=8)
plt.tight_layout()
plt.show()
````

## 7. Train Best Model on Full Training Set

````python
# ── Pick the model with highest F1-macro from CV ─────────────────────────────
best_model_name = max(cv_results, key=lambda k: cv_results[k]['f1_macro'])
print(f'Best model by F1-macro: {best_model_name}')

best_pipeline = pipelines[best_model_name]
best_pipeline.fit(X_train_text, y_train)
print('Model trained on full training set.')
````

## 8. Evaluation on Held-Out Test Set

````python
y_pred = best_pipeline.predict(X_test_text)

acc         = accuracy_score(y_test, y_pred)
f1_macro    = f1_score(y_test, y_pred, average='macro')
f1_weighted = f1_score(y_test, y_pred, average='weighted')

print(f'Test Accuracy        : {acc:.4f} ({acc*100:.2f}%)')
print(f'Test F1-macro        : {f1_macro:.4f}')
print(f'Test F1-weighted     : {f1_weighted:.4f}')
print()
print('── Classification Report ──')
print(classification_report(y_test, y_pred, digits=4))
````

````python
# ── Confusion Matrix ─────────────────────────────────────────────────────────
labels = sorted(set(y_test))
cm = confusion_matrix(y_test, y_pred, labels=labels)

fig, ax = plt.subplots(figsize=(7, 6))
sns.heatmap(
    cm, annot=True, fmt='d', cmap='Blues',
    xticklabels=labels, yticklabels=labels,
    linewidths=0.5, ax=ax
)
ax.set_xlabel('Predicted Label')
ax.set_ylabel('True Label')
ax.set_title(f'Confusion Matrix — {best_model_name}')
plt.tight_layout()
plt.show()
````

````python
# ── Per-class F1 bar chart ───────────────────────────────────────────────────
report = classification_report(y_test, y_pred, output_dict=True)
per_class_f1 = {k: v['f1-score'] for k, v in report.items() if k in labels}

fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(per_class_f1.keys(), per_class_f1.values(),
       color=sns.color_palette('muted', len(per_class_f1)))
ax.set_ylim(0, 1.05)
ax.set_ylabel('F1-Score')
ax.set_title('Per-Class F1-Score on Test Set')
for i, (k, v) in enumerate(per_class_f1.items()):
    ax.text(i, v + 0.01, f'{v:.3f}', ha='center', fontsize=10)
plt.tight_layout()
plt.show()
````

````python
# ── Misclassification analysis ───────────────────────────────────────────────
test_df['predicted'] = y_pred
errors = test_df[test_df['Label'] != test_df['predicted']]
print(f'Total misclassified: {len(errors)} / {len(test_df)} ({100*len(errors)/len(test_df):.1f}%)')
print('\nSample misclassifications:')
print(errors[['Transaction_Text', 'Label', 'predicted']]
      .sample(min(8, len(errors)), random_state=RANDOM_STATE)
      .to_string(index=False))
````

## 9. Save Model

````python
model_path = os.path.join(MODEL_DIR, 'transaction_classifier.pkl')
joblib.dump(best_pipeline, model_path)
print(f'Model saved → {model_path}')

# ── Quick sanity check on fresh load ────────────────────────────────────────
loaded_model = joblib.load(model_path)

test_transactions = [
    'Swiggy order payment | Ref:abc123 | Amount: INR 450.00',
    'HDFC home loan EMI | Ref:def456 | Amount: INR 25000.00',
    'Amazon online shopping | Ref:ghi789 | Amount: INR 1200.00',
    'Zerodha stock purchase | Ref:jkl012 | Amount: INR 50000.00',
    'Ola cab fare | Ref:mno345 | Amount: INR 300.00',
]

clean_tests = [clean_text(t) for t in test_transactions]
preds = loaded_model.predict(clean_tests)

print('\n── Inference on new transactions ──')
for tx, pred in zip(test_transactions, preds):
    print(f'  {tx[:55]}...  →  {pred}')
````
